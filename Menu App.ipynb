```python
import pandas as pd
import random
import time
```

```python
meals = pd.read_csv(r'C:\Users\saeth\OneDrive\Documents\Operation Get Employed\Code Challenges\menu app\meals.csv', header = 0)
meals
```

```python
class menu_builder():
    def __init__(self, meals):
        print('menu builder initialized')
    
    def select_breakfast(self, meals):
        breakfast = list(meals[meals['breakfast'] == 1].index)
        breakfast_idx = random.choices(breakfast, k = 7)
        breakfast_menu = meals.iloc[breakfast_idx, [0]]
        return list(breakfast_menu['meals'])
    
    def select_lunches(self,meals,dinner_menu, dinners_idx):
        n = 7-len(dinner_menu[dinner_menu['leftovers']==1])
        lunches = list(meals[meals['lunch'] == 1].index)
        lunches_idx = random.sample(lunches, n)
        dinner_with_leftovers = list(dinner_menu[dinner_menu['leftovers']==1].index)
        lunches_idx = lunches_idx + dinner_with_leftovers
        lunch_menu = meals.iloc[lunches_idx, [0]]
        return list(lunch_menu['meals'])
    
    def select_snacks(self, meals):
        snacks = list(meals[meals['snack'] == 1].index)
        snacks_idx = random.choices(snacks, k = 7)
        snack_menu = meals.iloc[snacks_idx, [0]]
        return list(snack_menu['meals'])
    
    def select_dinners(self, meals):
        dinners = list(meals[meals['dinner'] == 1].index)
        dinners_idx = random.sample(dinners, 7)
        dinner_menu = meals.iloc[dinners_idx, [0,5]]
        return dinner_menu, dinners_idx, list(dinner_menu['meals'])
    
    def build_menu(self, meals, weeks = 4):
        Menu = pd.DataFrame(pd.DataFrame({'week': [],
                              'day': [],
                              'breakfast': [],
                              'lunch': [],
                              'dinner': []}))
        DOW = ['MON', 'TUE', 'WED', 'THU', 'FRI', 'SAT', 'SUN']
        
        for i in range(1, weeks+1):
            week = ['week '+ str(i)]*7
            dinner_menu, dinners_idx, dinner_list = self.select_dinners(meals)
            lunch_list = self.select_lunches(meals, dinner_menu, dinners_idx)
            breakfast_list = self.select_breakfast(meals)
            snack_list = self.select_snacks(meals)
            
            week_menu = pd.DataFrame({'week': week,
                                      'day': DOW,
                                      'breakfast': breakfast_list,
                                      'lunch': lunch_list,
                                      'dinner': dinner_list})
            
            Menu = pd.concat([Menu,week_menu],ignore_index = True)
        Menu.to_csv("menu_"+(time.strftime("%Y%m%d")+".csv"), index = False)
            
        return Menu
        
```
